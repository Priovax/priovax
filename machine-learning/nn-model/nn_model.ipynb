```
import pandas as pd

df = pd.read_csv("/content/dataset.csv")
```

```
cat_names = ["chronic_disease", "gender_binary", "diabetes", "neuro", "hypertension", "cancer", "ortho",
                 "respiratory", "cardiacs", "kidney", "blood", "prostate", "thyroid"]

cont_names = ["age"]
```

```
splits = RandomSplitter(valid_pct=0.2)(range_of(df))
```

```
to = TabularPandas(df, procs=[Normalize, Categorify],
                   cat_names = cat_names,
                   cont_names = cont_names,
                   y_names='death_binary',
                   y_block = CategoryBlock,
                   splits=splits)

dls = to.dataloaders(bs=64)
dls.show_batch()
```

```
learn = tabular_learner(dls, metrics=[accuracy, Precision(average = "macro"), Recall(average = "macro")])
```

```
learn.fit_one_cycle(1)
learn.show_results()
```
